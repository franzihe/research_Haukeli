```python
import sys
sys.path.append('/uio/kant/geo-metos-u7/franzihe/Documents/Thesis/Python')


import netCDF4
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import createFolder as cF
```

```python
station       = 'Haukeliseter'
forecast_time = '00'
```

```python
savefig = 0
```

```python
acc_hours = 24
```

```python
ana_date = [201611, 201612, 201701, 201702]
```

```python
var_name = 'precipitation_amount_acc'
```

```python
if savefig == 1:
    figdir_var = '../../Figures/Weathermast_MEPS/%s_statistic' %(var_name)
    cF.createFolder('%s/' %(figdir_var))
form = 'png' 
```

```python
f = dict()
for i in range(0,4):
    f[ana_date[i]] = netCDF4.Dataset('../../Data/Weathermast_MEPS/%s/%s_%s.nc' 
                     %(station,ana_date[i],forecast_time),
                     'r')
```

```python
obs = dict()
MEPS = dict()
MEPS_EM09_mean = dict()
Difference = dict()
Difference_EM09 = dict()
Date = dict()
```

```python
for i in range(0,len(ana_date)):
    obs[ana_date[i]] = (f[ana_date[i]].variables['weathermast_'+var_name+'_48'][:])
    MEPS[ana_date[i]] = (f[ana_date[i]].variables[var_name][:])
    MEPS_EM09_mean[ana_date[i]] = np.nanmean(MEPS[ana_date[i]][:,:,:],axis=1)
    Difference[ana_date[i]] = f[ana_date[i]].variables[var_name+'_diff_perc'][:]
    Difference_EM09[ana_date[i]] = f[ana_date[i]].variables[var_name+'_ensemble_mean_diff_perc'][:,:]
    
    Date[ana_date[i]] = (f[ana_date[i]].variables['Date'][:,:])
    Date[ana_date[i]] = np.flipud(Date[ana_date[i]])
```

```python
R_labels = dict()
Table_values = dict()
```

```python
col_labels = ['obs [mm]', 'MEPS [mm]', 'Difference [%]', 'Average difference [%]']

for i in range(0,len(ana_date)):
    ## Nov 2016
    if ana_date[i] == 201611:
        dates = [12, 13, 16, 17, 18, 29]
        
    ## Dec 2016    
    elif ana_date[i] == 201612:
        dates = np.arange(21,27)
    
    ## Jan 2017
    elif ana_date[i] == 201701:
        dates = [3, 6, 9, 10, 11, 12, 13, 29]
    
    ## Feb 2017
    elif ana_date[i] == 201702:
        dates = np.arange(2,5)
        
    row_labels = []
    table_vals = []

    for ens_memb in range(0,10):
        for t in dates:
        #print( t)
            if t == dates[0]:
                row_labels.append('')
                row_labels.append('%s %s %s' %(Date[ana_date[i]][1,t-1],Date[ana_date[i]][2,t-1],Date[ana_date[i]][3,t-1]))
                table_vals.append(['','','','ensemble member %s' %ens_memb])
                table_vals.append([ np.round(       obs[ana_date[i]][acc_hours,t-1],2), 
                                np.round(      MEPS[ana_date[i]][acc_hours,ens_memb,t-1],2), 
                                np.round(Difference[ana_date[i]][acc_hours,ens_memb,t-1],2), 
                        ''])
            elif t == dates[-1]:
                row_labels.append('%s %s %s' %(Date[ana_date[i]][1,t-1],Date[ana_date[i]][2,t-1],Date[ana_date[i]][3,t-1]))
                row_labels.append('')
                table_vals.append([ np.round(       obs[ana_date[i]][acc_hours,t-1],2), 
                                np.round(      MEPS[ana_date[i]][acc_hours,ens_memb,t-1],2), 
                                np.round(Difference[ana_date[i]][acc_hours,ens_memb,t-1],2), 
                                np.round(np.nanmean([Difference[ana_date[i]][acc_hours,ens_memb,np.array(dates)-1]]),2)])
                table_vals.append(['', '', '', ''])
            else:
                row_labels.append('%s %s %s' %(Date[ana_date[i]][1,t-1],Date[ana_date[i]][2,t-1],Date[ana_date[i]][3,t-1]))
                table_vals.append([ np.round(       obs[ana_date[i]][acc_hours,t-1],2), 
                                np.round(      MEPS[ana_date[i]][acc_hours,ens_memb,t-1],2), 
                                np.round(Difference[ana_date[i]][acc_hours,ens_memb,t-1],2), 
                        ''])
        if ens_memb == 9:
            for t in dates:
        #print( t)
                if t == dates[0]:
                    row_labels.append('')
                    row_labels.append('%s %s %s' %(Date[ana_date[i]][1,t-1],Date[ana_date[i]][2,t-1],Date[ana_date[i]][3,t-1]))
                    table_vals.append(['','','','ensemble mean' ])
                    table_vals.append([ np.round(            obs[ana_date[i]][acc_hours,t-1],2), 
                                    np.round( MEPS_EM09_mean[ana_date[i]][acc_hours,t-1],2), 
                                    np.round(Difference_EM09[ana_date[i]][acc_hours,t-1],2), 
                            ''])
                elif t == dates[-1]:
                    row_labels.append('%s %s %s' %(Date[ana_date[i]][1,t-1],Date[ana_date[i]][2,t-1],Date[ana_date[i]][3,t-1]))
                    row_labels.append('')
                    table_vals.append([ np.round(            obs[ana_date[i]][acc_hours,t-1],2), 
                                    np.round( MEPS_EM09_mean[ana_date[i]][acc_hours,t-1],2), 
                                    np.round(Difference_EM09[ana_date[i]][acc_hours,t-1],2), 
                                    np.round(np.nanmean([Difference_EM09[ana_date[i]][acc_hours,np.array(dates)-1]]),2)])
                    table_vals.append(['', '', '', ''])
                else:
                    row_labels.append('%s %s %s' %(Date[ana_date[i]][1,t-1],Date[ana_date[i]][2,t-1],Date[ana_date[i]][3,t-1]))
                    table_vals.append([ np.round(            obs[ana_date[i]][acc_hours,t-1],2), 
                                    np.round( MEPS_EM09_mean[ana_date[i]][acc_hours,t-1],2), 
                                    np.round(Difference_EM09[ana_date[i]][acc_hours,t-1],2), 
                            ''])
        
        R_labels[ana_date[i]] = row_labels[:]
        Table_values[ana_date[i]] = table_vals[:]
```

```python
for i in range(0,len(ana_date)):
    fig = plt.figure()
    ax = fig.add_subplot(111)

# Draw table
    the_table = plt.table(cellText=Table_values[ana_date[i]],
                      colWidths=[0.25] * len(col_labels),
                      rowLabels=R_labels[ana_date[i]],
                      colLabels=col_labels,
                      loc='center')
    the_table.auto_set_font_size(False)
    the_table.set_fontsize(24)
    the_table.scale(4, 4)

# Removing ticks and spines enables you to get the figure only with table
    plt.tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=False)
    plt.tick_params(axis='y', which='both', right=False, left=False, labelleft=False)
    for pos in ['right','top','bottom','left']:
        plt.gca().spines[pos].set_visible(False)
    if ana_date[i] == 201611 or ana_date[i] == 201612:
        yy = 10.38
    elif ana_date[i] == 201701:
        yy = 12.78
    elif ana_date[i] == 201702:
        yy = 6.78
    fig_name = str(ana_date[i])+'_'+str(acc_hours)+'.'+form
    
    ax.set_title('%s %s -- %sh accumulation' %(Date[ana_date[i]][2,t-1],Date[ana_date[i]][3,t-1],acc_hours),y=yy, 
             fontsize=24)
#ax.title.set_text('First Plot')
#plt.savefig('matplotlib-table.png', bbox_inches='tight', pad_inches=0.05)

    if savefig == 1:
        plt.savefig('%s/%s' % (figdir_var, fig_name), format = form, bbox_inches='tight')
        print('plot saved: %s/%s' %(figdir_var,fig_name))
    plt.close()
```

```python

OOD = dict()
obs_month_sum = []

MOD = dict()
MEPS_month_sum = []

for i in range(0,len(ana_date)):
    ## Nov 2016
    if ana_date[i] == 201611:
        dates = [12, 13, 16, 17, 18, 29]
        
    ## Dec 2016    
    elif ana_date[i] == 201612:
        dates = np.arange(21,27)
    
    ## Jan 2017
    elif ana_date[i] == 201701:
        dates = [3, 6, 9, 10, 11, acc_hours, 13, 29]
    
    ## Feb 2017
    elif ana_date[i] == 201702:
        dates = np.arange(2,5)
    
    obs_of_dates = []
    MEPS_of_dates = []
    for t in dates:
        obs_of_dates.append(obs[ana_date[i]][acc_hours,t-1])
        MEPS_of_dates.append(MEPS_EM09_mean[ana_date[i]][acc_hours,t-1])
    OOD[ana_date[i]] = obs_of_dates[:]
    MOD[ana_date[i]] = MEPS_of_dates[:]                         
    obs_month_sum.append(np.nansum(OOD[ana_date[i]]))
    MEPS_month_sum.append(np.nansum(MOD[ana_date[i]]))
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
#y = [1, 2, 3, 4, 5, 4, 3, 2, 1, 1, 1, 1, 1, 1, 1, 1]    
#col_labels = ['col1', 'col2', 'col3']
col_labels = ['obs [mm]', 'MEPS [mm]', 'Difference [%]', 'Average difference [%]']

#row_labels = ['row1', 'row2', 'row3']
#row_labels = [np.array(R_labels[201611])[-8:],
 #             np.array(R_labels[201612])[-8:],
  #            np.array(R_labels[201701])[-10:],
   #           np.array(R_labels[201702])[-5:] ]
row_labels = np.concatenate((np.array(R_labels[201611])[-8:], np.array(R_labels[201612])[-8:],
                            np.array(R_labels[201701])[-10:], np.array(R_labels[201702])[-5:],
                            [''], ['Winter 16/17']),axis=0)
table_vals = np.concatenate( (np.array(Table_values[201611])[-8:,:],
              np.array(Table_values[201612])[-8:,:],
              np.array(Table_values[201701])[-10:,:],
              np.array(Table_values[201702])[-5:,:]), axis=0)
table_vals = np.append(table_vals,[['', '','','']],axis=0)
table_vals = np.append(table_vals,[[np.round(np.sum(obs_month_sum),2), 
                   np.round(np.sum(MEPS_month_sum),2), 
                  np.round(((np.sum(MEPS_month_sum) - np.sum(obs_month_sum))/np.sum(obs_month_sum))*100,2), 
               '']],axis=0)






# Draw table
the_table = plt.table(cellText=table_vals,
                #      colWidths=[0.25] * len(col_labels),
                      rowLabels=row_labels,
                      colLabels=col_labels,
                      loc='center')
the_table.auto_set_font_size(False)
the_table.set_fontsize(24)
the_table.scale(4, 4)

# Removing ticks and spines enables you to get the figure only with table
plt.tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=False)
plt.tick_params(axis='y', which='both', right=False, left=False, labelleft=False)
for pos in ['right','top','bottom','left']:
    plt.gca().spines[pos].set_visible(False)

fig_name = 'Winter20162017_'+str(acc_hours)+'_ensemble_mean.'+form

yy=4.28
ax.set_title('Winter 2016/2017 -- %sh accumulation -- ensemble mean' %(acc_hours),y=yy, 
             fontsize=24)
#ax.title.set_text('First Plot')
#plt.savefig('matplotlib-table.png', bbox_inches='tight', pad_inches=0.05)

if savefig == 1:
    plt.savefig('%s/%s' % (figdir_var, fig_name), format = form, bbox_inches='tight')
    print('plot saved: %s/%s' %(figdir_var,fig_name))
else:
    plt.show()
plt.close()
```

```python
row_labels = []
table_vals = []

for ens_memb in range(0,10):
    for i in range(0,len(ana_date)):
    ## Nov 2016
        if ana_date[i] == 201611:
            dates = [12, 13, 16, 17, 18, 29]
        
    ## Dec 2016    
        elif ana_date[i] == 201612:
            dates = np.arange(21,27)
    
    ## Jan 2017
        elif ana_date[i] == 201701:
            dates = [3, 6, 9, 10, 11, 12, 13, 29]
    
    ## Feb 2017
        elif ana_date[i] == 201702:
            dates = np.arange(2,5)
        
        obs_month = np.nansum(     [np.nansum(obs[201611][acc_hours,[11,12,15,16,17,28]]),
                                np.nansum(obs[201612][acc_hours,20:26]),
                                np.nansum(obs[201701][acc_hours,[2,5,8,9,10,11,12,28]]),
                                np.nansum(obs[201702][acc_hours,[1,2,3]])])
    
        MEPS_month  = np.nansum(   [np.nansum(MEPS[201611][acc_hours,ens_memb,[11,12,15,16,17,28]]) , 
                                np.nansum(MEPS[201612][acc_hours,ens_memb,20:26]),
                                np.nansum(MEPS[201701][acc_hours,ens_memb,[2,5,8,9,10,11,12,28]]),
                                np.nansum(MEPS[201702][acc_hours,ens_memb,[1,2,3]])])
        Diff_year = ((MEPS_month-obs_month)/obs_month)*100
    

        if ana_date[i] == 201611:
            table_vals.append(['','','','ensemble member %s' %ens_memb])
        table_vals.append([np.round(np.nansum(obs[ana_date[i]][acc_hours,np.array(dates)-1]),2), 
                           np.round(np.nansum(MEPS[ana_date[i]][acc_hours,ens_memb,np.array(dates)-1]),2), 
                           np.round((np.nansum(MEPS[ana_date[i]][acc_hours,ens_memb,np.array(dates)-1]) - np.nansum(obs[ana_date[i]][acc_hours,np.array(dates)-1]))/np.nansum(obs[ana_date[i]][acc_hours,np.array(dates)-1])*100,2), 
                               ''])
        if ana_date[i] == 201702:
            table_vals.append([np.round(obs_month,2), np.round(MEPS_month,2),  np.round(Diff_year,2), ''])
            table_vals.append(['','','',''])
    

        row_labels = np.append((row_labels),['%s %s' %(Date[ana_date[i]][2,t-1],Date[ana_date[i]][3,t-1])])
    if ens_memb == 9:
        row_labels = np.append(row_labels,(['Winter 2016/17'], ['']))
    else:
        row_labels = np.append(row_labels,(['Winter 2016/17'], [''], ['']))
row_labels = np.append([''], row_labels)
               
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
col_labels = ['obs [mm]', 'MEPS [mm]', 'Difference [%]', 'Ensemble member']


# Draw table
the_table = plt.table(cellText=table_vals,
                      colWidths=[0.25] * len(col_labels),
                      rowLabels=row_labels,
                      colLabels=col_labels,
                      loc='center')
the_table.auto_set_font_size(False)
the_table.set_fontsize(24)
the_table.scale(4, 4)

# Removing ticks and spines enables you to get the figure only with table
plt.tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=False)
plt.tick_params(axis='y', which='both', right=False, left=False, labelleft=False)
for pos in ['right','top','bottom','left']:
    plt.gca().spines[pos].set_visible(False)

fig_name = 'Winter20162017_'+str(acc_hours)+'_EM09.'+form

yy=8.38
ax.set_title('Winter 2016/2017 -- %sh accumulation' %(acc_hours),y=yy, 
             fontsize=24)
#ax.title.set_text('First Plot')
#plt.savefig('matplotlib-table.png', bbox_inches='tight', pad_inches=0.05)

if savefig == 1:
    plt.savefig('%s/%s' % (figdir_var, fig_name), format = form, bbox_inches='tight')
    print('plot saved: %s/%s' %(figdir_var,fig_name))
else:
    plt.show()
plt.close()
```
